```python
# Import pandas and matplotlib.pyplot
%matplotlib inline
import pandas as pd
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

```

```python
# Read in the csv as a DataFrame
office_df = pd.read_csv('datasets/office_episodes.csv', parse_dates=['release_date'])
```

```python
# Initiatlize two empty lists
cols = []
sizes = []

#color lavels
col_labels =[]

# Iterate through the DataFrame, and assign colors based on the rating
for ind, row in office_df.iterrows():
    if row['scaled_ratings'] < 0.25:
        cols.append('red')
        col_labels.append('Low Rating')
    elif row['scaled_ratings'] < 0.50:
        cols.append('orange')
        col_labels.append('Average Rating')
    elif row['scaled_ratings'] < 0.75:
        cols.append('lightgreen')
        col_labels.append('High Rating')
    else:
        cols.append('darkgreen')
        col_labels.append('Very High Rating')
```

```python
colors_lables = zip(cols, col_labels)
```

```python
colors_lables = list(set(colors_lables))
lables = [lable for color,lable in colors_lables]
```

```python
colors_lables
```

```python
lables_patchs = []
#Uses mpatches package to generate a name for each unique item in the color_labels list. 
for item in colors_lables:
    add_patch = mpatches.Patch(color=item[0], label=item[1])
    lables_patchs.append(add_patch)
```

```python
# Iterate through the DataFrame, and assign a size based on whether it has guests        
for ind, row in office_df.iterrows():
    if row['has_guests'] == False:
        sizes.append(25)
    else:
        sizes.append(250)
```

```python
# For ease of plotting, add our lists as columns to the DataFrame
office_df['colors'] = cols
office_df['sizes'] = sizes
```

```python
# Set the figure size and plot style        
plt.rcParams['figure.figsize'] = [11, 7]
plt.style.use('fivethirtyeight')

# Create the figure
fig = plt.figure()

# Create a scatter plot with the episode number on the x axis, and the viewership on the y axis
plt.scatter(x=office_df.episode_number, y=office_df.viewership_mil, \
                 # Assign our color list as the colors
                 c=cols, \
                 # Assign our size lists as the sizes
                 s=sizes)
# Create a title
plt.title("Popularity, Quality, and Guest Appearances on the Office", fontsize=28)

# Create an x-axis label
plt.xlabel("Episode Number", fontsize=18)

# Create a y-axis label
plt.ylabel("Viewership (Millions)", fontsize=18)

plt.legend(lables_patchs, lables)

# Show the plot
plt.show()
```
